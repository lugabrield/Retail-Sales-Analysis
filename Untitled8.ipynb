```python
import pandas as pd
import seaborn as sns
```

```python
varejo = pd.read_excel('/content/varejo.xlsx')
varejo.head()
```

```python
varejo.info()
```

```python
varejo.groupby('idcanalvenda').idcompra.nunique()
```

```python
varejo['idcanalvenda'] = varejo['idcanalvenda'].str.replace('APP','Aplicativo')
varejo.groupby('idcanalvenda').idcompra.nunique()
```

```python
varejo.groupby('bandeira').idcompra.nunique()
```

```python
varejo.groupby('Nome_Departamento').idcompra.nunique()
```

```python
varejo.groupby('estado').idcompra.nunique()
```

```python
varejo['Nome_Departamento'] = varejo['Nome_Departamento'].str.replace(' ','_')
varejo.groupby('Nome_Departamento').idcompra.nunique()
```

```python
varejo.info()
```

```python
varejo[varejo['estado'].isnull()]
```

```python
varejo_sem_nulos = varejo.dropna()
varejo_sem_nulos.info()
```

```python
varejo['estado'].fillna('MS',inplace=True) # É igual a varejo['estado'] =  varejo['estado'].fillna('MS')
varejo.info()
```

```python
media_preco = varejo['Preço'].mean()
varejo['Preço'].fillna(media_preco,inplace=True)

```

```python
media_preco
```

```python
varejo.info()
```

```python
varejo[['Preço','Preço_com_frete']].describe()
```

```python
preco_errado = varejo.query('Preço > Preço_com_frete')

preco_correto = varejo.query('Preço < Preço_com_frete')
```

```python
preco_correto[['Preço','Preço_com_frete']].describe()
```

```python
preco_correto.query('Nome_Departamento == "Esporte_e_Lazer" and estado == "SP"').describe()
```

```python
preco_correto.groupby('Nome_Departamento').idcompra.nunique().sort_values(ascending = False).reset_index()
```

```python
round(preco_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index(),2)
```

```python
preco_correto.groupby('Data').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
preco_correto['mes'] = preco_correto['Data'].dt.month
preco_correto.head()
```

```python
preco_correto.groupby('mes').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
cliente = pd.read_excel('/content/cliente_varejo.xlsx')
```

```python
cliente.head()
```

```python
preco_correto.head()
```

```python
cliente[['idade','renda']].describe()
```

```python
cliente.info()
```

```python
cliente = cliente.astype({'renda':'float'})
```

```python
cliente.info()
```

```python
varejo_cliente = preco_correto.merge(cliente, how = 'left', on = 'cliente_Log')
varejo_cliente.head()
```

```python
agg_idcanal_renda = round(varejo_cliente.groupby('idcanalvenda')['renda'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_idcanal_renda
```

```python
agg_idade_bandeira = round(varejo_cliente.groupby('bandeira')['idade'].agg('mean').sort_values(ascending=False).reset_index(),0)
agg_idade_bandeira
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(7,4))
plt.bar(agg_idade_bandeira['bandeira'],agg_idade_bandeira['idade'], color ='green')
plt.ylabel('Média de idade')
plt.title('Idade média por bandeira',loc='center',fontsize='18',fontweight='bold',color='black');
```

```python
fig = plt.figure(figsize =(7,4))
plt.bar(agg_idcanal_renda['idcanalvenda'],agg_idcanal_renda['renda'],color = 'green')
plt.ylabel('Renda média')
plt.title('Renda média por canal',loc ='center',fontsize='18',fontweight='bold',color='black');
```

```python
venda_por_data = preco_correto.groupby('Data').idcompra.nunique().reset_index()
venda_por_data
```

```python
fig = plt.figure(figsize =(12,4))
plt.xlabel('Data da venda')
plt.ylabel('Qtd de vendas')
x = venda_por_data['Data']
y = venda_por_data['idcompra']
plt.plot(x,y,color = 'green',linewidth='2')
plt.title('Qtd de vendas por data',loc='center',fontsize='18',fontweight='bold',color='black')
plt.show()
```

```python
nome_departamento_preco_com_frete = round(preco_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index(),2)
nome_departamento_preco_com_frete
```

```python
fig = plt.figure(figsize=(20,14))
x = nome_departamento_preco_com_frete['Nome_Departamento']
y = nome_departamento_preco_com_frete['Preço_com_frete']
plt.bar(x,y,color='green')
plt.ylabel('Preço médio')
plt.title('Preço com frete por Departamento',loc='center',fontsize='18',fontweight='bold',color='black');
```
